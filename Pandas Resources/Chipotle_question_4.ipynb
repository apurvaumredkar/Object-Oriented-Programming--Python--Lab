### Step 4. How many products cost more than $10.00?

```python
chipo.groupby(['item_name'])[['item_price']].first().reset_index()
```

### Step 5. What is the price of each item?

```python
chipo[['item_name','item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name').reset_index()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
c = chipo['item_price'].str.split('$',expand=True)[1].astype('float32').idxmax()
chipo.loc[c][['item_name','item_price','quantity']]
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl']['quantity'].count()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
c = chipo[chipo['item_name'] == 'Canned Soda']
c[c['quantity']>1]['item_name'].count()
```
